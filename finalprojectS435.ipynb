```python
#Twitter
import requests
url = 'https://twitter.com/NCAA' 
data = requests.get(url) 
print(data.text)
```

```python
#extracting tweets#
all_tweets = []
html = BeautifulSoup(data.text, 'html.parser')
timeline = html.select('#timeline li.stream-item')
for tweet in timeline:
    tweet_id = tweet['data-item-id']
    tweet_text = tweet.select('p.tweet-text')[0].get_text()
    all_tweets.append({"id": tweet_id, "text": tweet_text}) 
```

```python
def correlation_matrix(finaldfv1):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm

    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 300)
    cax = ax1.imshow(finaldfv1.corr(), interpolation="nearest", cmap="hot")
    ax1.grid(True)
    plt.title('Correlation')
    labels=['Mens Team Average Annual Institutional Salary per FTE',        'Mens Team Athletic Student Aid','Total Revenues Mens Team and Men Coed Team','Mens Team Average Annual Institutional Salary per Head Coach']
    ax1.set_xticklabels(labels,fontsize=3)
    ax1.set_yticklabels(labels,fontsize=12)
    fig.colorbar(cax, ticks=[.1,1])
    plt.show()

correlation_matrix(finaldfv1)
```
